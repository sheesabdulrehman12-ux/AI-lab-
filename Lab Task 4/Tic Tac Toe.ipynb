```python
class DynamicTicTacToe:

    def __init__(self, size=3):
        self.size = size
        self.board = [[str(i * size + j + 1) for j in range(size)] for i in range(size)]
        self.current_player = "X"

    def display_board(self):
        print("\n" + "-" * (self.size * 4))
        for row in self.board:
            print(" | ".join(row))
            print("-" * (self.size * 4))
        print()

    def switch_player(self):
        if self.current_player == "X":
            self.current_player = "O"
        else:
            self.current_player = "X"

    def make_move(self, move):
        for i in range(self.size):
            for j in range(self.size):
                if self.board[i][j] == str(move):
                    self.board[i][j] = self.current_player
                    return True
        return False

    def check_winner(self):
        b = self.board
        s = self.size

        for i in range(s):
            if all(b[i][j] == b[i][0] for j in range(s)):
                return True

        for j in range(s):
            if all(b[i][j] == b[0][j] for i in range(s)):
                return True

        if all(b[i][i] == b[0][0] for i in range(s)):
            return True

        if all(b[i][s - 1 - i] == b[0][s - 1] for i in range(s)):
            return True

        return False

    def is_draw(self):

        for row in self.board:
            for cell in row:
                if cell not in ["X", "O"]:
                    return False
        return True

    def play(self):

        print(f"\nWelcome to {self.size}x{self.size} Tic Tac Toe!\n")

        while True:
            self.display_board()
            print(f"Player {self.current_player}'s turn")

            try:
                move = int(input("Enter your move number: "))
            except ValueError:
                print("Please enter a valid number.")
                continue

            if not self.make_move(move):
                print("Invalid move! Try again.")
                continue

            if self.check_winner():
                self.display_board()
                print(f"Player {self.current_player} wins!")
                break

            if self.is_draw():
                self.display_board()
                print("It's a draw!")
                break

            self.switch_player()

try:
    n = int(input("Enter board size (minimum 3): "))
    if n < 3:
        print("Board size must be 3 or greater.")
    else:
        game = DynamicTicTacToe(size=n)
        game.play()
except ValueError:
    print("Invalid board size. Please enter a number")
```
