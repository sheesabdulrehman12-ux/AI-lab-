```python
import pandas as pd
import numpy as np
```

## 1. Reading CSV File

```python
# Replace 'dataset.csv' with your Kaggle dataset filename
data = pd.read_csv('dataset.csv')
data.head()
```

## 2. Printing Rows and Columns / Shape of DataFrame

```python
print('We have {} rows.'.format(data.shape[0]))
print('We have {} columns.'.format(data.shape[1]))
```

## 3. Checking Null Values

```python
np.sum(pd.isnull(data))
```

## 4. Checking Unique Values of a Specific Column

```python
# Replace 'columnname' with your actual column name
data['columnname'].unique()
```

## 5. Filling Null Values

```python
# Replace 'columnname' with your actual column name
num = data['columnname'].mode()[0]
data['columnname'] = data['columnname'].fillna(num)
data.head()
```

## 6. Type Conversion to Integer

```python
# Replace 'columnname' with your actual column name
data['columnname'] = data['columnname'].astype(np.int64)
data.dtypes
```

## 7. Dropping Unnecessary Columns

```python
# Replace 'id' with your unnecessary column name
data.drop('id', axis=1, inplace=True)
data.head()
```

## 8. Checking Datatypes of All Columns

```python
data.dtypes
```

## 9. Splitting Dataset into X and Y

```python
x = data.iloc[:, 0:-1]
print(x.shape)

y = data.iloc[:, -1]
print(y.shape)
```

## 10. Converting Object Columns into Integer Columns

```python
cat_columns = x.select_dtypes(['object']).columns
x[cat_columns] = x[cat_columns].apply(lambda col: pd.factorize(col)[0])
x.head()
```
