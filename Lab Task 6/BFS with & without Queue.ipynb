```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': ['G', 'H'],
    'E': [],
    'F': ['I', 'K'],
    'G': [],
    'H': ['L'],
    'I': [],
    'K': ['M'],
    'L': [],
    'M': []
}

def bfswithoutqueue(graph, start):
    visited = []         
    current_level = [start]

    while current_level:
        next_level = []   
        for node in current_level:
            if node not in visited:
                print(node)   
                visited.append(node)
                for neighbour in graph[node]:
                    next_level.append(neighbour)
        current_level = next_level

print("BFS Traversal (Without Queue):")
bfswithoutqueue(graph, 'A')
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': ['G', 'H'],
    'E': [],
    'F': ['I', 'K'],
    'G': [],
    'H': ['L'],
    'I': [],
    'K': ['M'],
    'L': [],
    'M': []
}

def bfs_with_queue(graph, start):
    visited = []       
    queue = [start]     

    while queue:
        node = queue.pop(0)  
        if node not in visited:
            print(node)     
            visited.append(node)
            for neighbour in graph[node]:
                queue.append(neighbour)

print("BFS Traversal (With Queue):")
bfs_with_queue(graph, 'A')
```
