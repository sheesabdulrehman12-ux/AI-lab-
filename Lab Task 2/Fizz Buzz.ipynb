```python
for comp_num in range(1, 101):  
    print(f"\nComputer's number = {comp_num}")

    try:
        user_num = int(input("User -> Enter your number: "))
    except ValueError:
        print("❌ Invalid input! Please enter a valid integer.")
        break  

    total = comp_num + user_num
    print(f"Sum = {comp_num} + {user_num} = {total}")

    if total % 3 == 0 and total % 5 == 0:
        print("FizzBuzz")
    elif total % 3 == 0:
        print("Fizz")
    elif total % 5 == 0:
        print("Buzz")
    else:
        print("No FizzBuzz match!")
```
