```python
def dfs_stack(graph, start):
    visited = []
    stack = [start]

    print("DFS Traversal Using Stack:")

    while stack:
        node = stack.pop()
        if node not in visited:
            print(node, end=" ")
            visited.append(node)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    stack.append(neighbor)


graph = {
    0: [1, 2],
    1: [3, 4],
    2: [5],
    3: [],
    4: [5],
    5: []
}

dfs_stack(graph, 0)
```
