```python
class ReflexAgent:
    def __init__(self, desired_temp):
        self.desired_temp = desired_temp
        self.room_history = {}

    def sense_environment(self, room, current_temp):
        self.room = room
        self.current_temp = current_temp
        if room not in self.room_history:
            self.room_history[room] = "AC OFF"

    def analyze_condition(self):
        last_action = self.room_history[self.room]
        decision = "No Change"

        if self.current_temp > self.desired_temp + 1 and last_action != "AC ON":
            decision = "AC Turned ON"
            self.room_history[self.room] = "AC ON"
        elif self.current_temp <= self.desired_temp - 1 and last_action != "AC OFF":
            decision = "AC Turned OFF"
            self.room_history[self.room] = "AC OFF"

        return decision

    def execute_action(self):
        result = self.analyze_condition()
        print(f"{self.room}: {self.current_temp} {result}")

room_temps = {
    "Living Room": 20,
    "Bedroom": 21,
    "Dining Room": 23,
    "Kitchen": 25,
    "Washroom": 22
}

agent = ReflexAgent(22)

print("Initial Room Temperature Check:")
for room, temp in room_temps.items():
    agent.sense_environment(room, temp)
    agent.execute_action()

updated_temps = {
    "Living Room": 24,
    "Bedroom": 20,
    "Dining Room": 21,
    "Kitchen": 23,
    "Washroom": 25
}

print("\nSecond Room Temperature Check:")
for room, temp in updated_temps.items():
    agent.sense_environment(room, temp)
    agent.execute_action()
```
